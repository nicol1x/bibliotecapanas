```python
import pandas as pd
import seaborn as sn
```

```python
vendas = pd.read_excel("varejo.xlsx")
vendas.info()
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP", "Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas.groupby("bandeira").idcompra.nunique()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ","_")
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas[vendas["estado"].isnull()]
```

```python
vendas["estado"].fillna("MS", inplace=True)
vendas.info()
```

```python
mediapreco = vendas["Preço"].mean()
vendas["Preço"].fillna(mediapreco, inplace =True)
vendas.info()
```

```python
mediana = vendas["Preço"].median()
mediana
```

```python
vendas[["Preço", "Preço_com_frete"]].describe()
```

```python
precoerrado = vendas.query("Preço > Preço_com_frete")
precocorreto = vendas.query("Preço < Preço_com_frete")
precocorreto.describe()
```

```python
precocorreto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
round(precocorreto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
```

```python
precocorreto.groupby("Data").idcompra.nunique().sort_values(ascending=False)
```

```python
precocorreto["mes"] = precocorreto["Data"].dt.month
precocorreto
```

```python
precocorreto.groupby("mes").idcompra.nunique().sort_values(ascending = False)
```

```python
cliente = pd.read_excel("/content/cliente_varejo.xlsx")
cliente.head()
```

```python
cliente.info()
```

```python
cliente[["idade", "renda"]].describe()
```

```python
vendascliente = precocorreto.merge(cliente,how = "left", on = "cliente_Log")
vendascliente
```

```python
aggidcanalrenda = round(vendascliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending = False).reset_index(),2)
aggidcanalrenda
```

```python
aggidadebandeira = round(vendascliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending = False).reset_index(),2)
aggidadebandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,4))
plt.bar(aggidadebandeira["bandeira"],aggidadebandeira["idade"], color = "green", )
plt.ylabel("Média Idade")
plt.title("Idade média por bandeira", loc = "center", fontsize = 18, fontweight = "bold", color = "green")
```

```python
fig2 = plt.figure(figsize = (7,4))
plt.bar(aggidcanalrenda["idcanalvenda"], aggidcanalrenda["renda"], color = "red")
plt.ylabel("Média renda")
plt.title("Renda média por canal de venda", loc= "center", fontsize = 18, fontweight = "bold", color = "red")
```

```python
vendaspordata = precocorreto.groupby("Data").idcompra.nunique().reset_index()
vendaspordata
```

```python
plt.figure(figsize = (15,8))

plt.xlabel("Data da venda")
plt.ylabel("Qtde de venda")
x = vendaspordata["Data"]
y = vendaspordata["idcompra"]

plt.plot(x,y, color =  "blue", linewidth = "2")
plt.show()
```

```python
aggdeptopreco = round(precocorreto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
aggdeptopreco
```

```python
fig3 = plt.figure(figsize = (15,4))
plt.bar(aggdeptopreco["Nome_Departamento"], aggdeptopreco["Preço_com_frete"], color =  "green")
plt.ylabel("Preço médio com frete")
plt.title("Preço médio com frete por departamento")
```
